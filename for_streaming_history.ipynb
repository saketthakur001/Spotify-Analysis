```python
# sort by date
df = df.sort_values("endTime")

# get song uri by song name and artist
def get_song_uri(song_name, artist):
    results = sp.search(q=f"track:{song_name} artist:{artist}", type="track")
    try:
        return results['tracks']['items'][0]['uri']
    except:
        return None

times_listened = df.trackName.value_counts()

# get songs that are listened x times
def get_songs_listened_x_times(x):
    name_and_artist = {"name":[], "artist":[], "date":[]}

    #if listened x times
    for i in times_listened.index:
        if times_listened[i] == x:
            name_and_artist["name"].append(i)
            name_and_artist["artist"].append(df[df.trackName == i].artistName.iloc[0])
            name_and_artist["date"].append(df[df.trackName == i].endTime.iloc[0])
    return name_and_artist

twice_listened_songs = get_songs_listened_x_times(x=3)
# arrage songs by date
twice_listened_songs = pandas.DataFrame(twice_listened_songs)
twice_listened_songs = twice_listened_songs.sort_values("date")
#reverse the order
twice_listened_songs = twice_listened_songs.iloc[::-1]
# print(x)
```
